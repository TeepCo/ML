```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np

import pandas as pd

def get_data(file):
    d = pd.read_csv(file,
                    index_col="beat",
                    usecols=["beat", "timestamp_m", "locLat", "locLon",
                             "verticalDistance", "horizontalDistance"],
                    converters=dict(beat=lambda x: pd.to_datetime(x, unit="s"),
                                    verticalDistance=lambda x: int(float(x)),
                                    horizontalDistance=lambda x: int(float(x))
                                    ))
    d["hour"] = (d["timestamp_m"] % 86_400_000) / 3_600_000
    d["locLon"] -= 14
    d["locLat"] -= 50
    return d

class AnimatedScatter(object):
    """An animated scatter plot using matplotlib.animations.FuncAnimation."""
    def __init__(self):
        self.dframe = get_data("/home/fiser/PycharmProjects/tacr/output/onesec_02_2020-12-09.csv") 

        # Setup the figure and axes...
        self.fig, self.ax = plt.subplots()
        # Then setup FuncAnimation.
        self.ani = animation.FuncAnimation(self.fig, self.update, interval=100, 
                                           frames=len(self.dframe)//100,
                                          init_func=self.setup_plot, blit=True)

    def setup_plot(self):
        """Initial drawing of the scatter plot."""
        self.scat = self.ax.scatter(x=[], y=[])
        self.ax.axis([0.0650, 0.0750, 0.525, 0.535])
        return self.scat,


    def update(self, i):
        """Update the scatter plot."""
        # Set x and y data...                          
        self.scat.set_offsets(self.dframe.iloc[:i*100, [2,1]].to_numpy())
        return self.scat,
```

```python
from IPython.display import HTML

anim = AnimatedScatter()
HTML(anim.ani.to_jshtml())
```
