```python
import folium
import pandas as pd
import matplotlib

d = pd.read_csv("/home/fiser/PycharmProjects/tacr/output/onesec_02_2020-12-08.csv")

m = folium.Map([50.5296678,14.070879], zoom_start=17)
limit = 80

for i,r in d.iterrows():
    location = (r["locLat"], r["locLon"])
    vdist = int(r["verticalDistance"])
    if vdist > limit:
        folium.Circle(location=location, radius=1,
                      color=["black","blue", "magenta", "red"][(vdist-limit)//80],
        fill=False).add_to(m)
```
